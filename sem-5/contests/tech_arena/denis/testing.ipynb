```python
datas = {}
for i in range(1,9):
    pdA = pd.read_csv(f'../data/train/problem-A/0{i}.csv')
    pdAans = pd.read_csv(f'../data/train/problem-A/0{i}.a.csv')
    for line in range(len(pdA)):
        inst = pdA['INST_RETIRED'][line]
        ips = pdAans['IPS'][line]
        time_get = inst/ips
        cluster = pdA['CLUSTER'][line]
        cpu = pdA['CPU_FREQ'][line]
        ddr = pdA['DDR_FREQ'][line]
        was = datas.get((cluster, cpu, ddr), [])
        was.append(time_get)
        datas[(cluster, cpu, ddr)] = was
```

```python
new_pda = []
for line in range(len(pdA)):
    cluster = pdA['CLUSTER'][line]
    cpu = pdA['CPU_FREQ'][line]
    ddr = pdA['DDR_FREQ'][line]
    inst = pdA['INST_RETIRED'][line]
    time = statistics.mode(datas[(cluster, cpu, ddr)])
    ips = inst/time
    new_pda.append({'IPS': ips, 'CLUSTER': cluster, 'CPU_FREQ': cpu, 'DDR_FREQ': ddr})
```

```python
def get_model_predict_by_args():
    train = pd.read_csv('../data/train/problem-B-C/01.csv')
    for i in range(2,37+1):
        path = f'../data/train/problem-B-C/{i//10}{i%10}.csv'
        train = pd.concat([train, pd.read_csv(path)])

    # train, eval= train_test_split(train, test_size=0.1)
    # target_col = 'IPC'
    # eval_x, eval_y = eval.drop(target_col, axis=1), eval[target_col]
    common_features = list(train.columns)
    common_features.remove('IPC')
    y_train=train['IPC']
    X_train=train[common_features]

    # param = [iter for iter in range(4, 30)]
    # for i in param:
    model = xgb.XGBRegressor()
    model.fit(X_train, y_train)
    return model
```

```python
def predict_args_for(path):
    df = pd.read_csv(path)
    
    common_features = list(df.columns)[3:6]
    common_features = ['INST_RETIRED']
    model = xgb.XGBRegressor()
    x_train = df[['CLUSTER', 'CPU_FREQ', 'DDR_FREQ']]
    y_train = df[common_features]
    model.fit(x_train, y_train)
    df_all = pd.read_csv('../configs.csv')
    x_test = df_all[['CLUSTER', 'CPU_FREQ', 'DDR_FREQ']]
    y_test = model.predict(x_test) 
    new_d = {}
    # for i, common_feature in enumerate(common_features):
    new_d['INST_RETIRED'] = [el[0] for el in y_test]
    for key in new_d:
        df_all[key] = new_d[key]
    return df_all
```

```python
model_all = get_model_predict_by_args()
```

```python
def get_result(model_all, df, path):
    result = model_all.predict(df)
    df['IPC'] = result * 2.55
    df.hist(bins=50, figsize=(15, 6))
    answer = df[['CLUSTER', 'CPU_FREQ', 'DDR_FREQ', 'IPC']]
    records = answer.to_dict('records')

    test = pd.read_csv(path)

    unused = []
    for line in range(len(test)):
        cluster = test['CLUSTER'][line]
        cpu = test['CPU_FREQ'][line]
        ddr = test['DDR_FREQ'][line]
        unused.append((cluster, cpu, ddr))

    new_records = []
    cnt = 0
    for record in records:
        if (record['CLUSTER'], record['CPU_FREQ'], record['DDR_FREQ']) in unused:
            cnt += 1
            continue
        new_records.append(record)
    import csv
    with open('people.csv', 'w', newline='') as output_file:
        dict_writer = csv.DictWriter(output_file, list(records[0].keys()))
        dict_writer.writeheader()
        dict_writer.writerows(new_records)
```

```python
path = '../data/test/problem-C/C2.csv'
df = predict_args_for(path)
get_result(model_all, df, path)
```
