```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score, roc_curve, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay, balanced_accuracy_score
```

```python
def read_a(number):
    return pd.read_csv(f'./data/train/problem-A/0{number}.a.csv')

df_a_01 = read_a(1)
df_a_01.info()
```

```python
df_a_all =  read_a(1)
for i in range(2, 9):
    df_a_i = read_a(i)
    df_a_all = pd.concat([df_a_all, df_a_i])

print(df_a_all.info())
df_a_all.head()
```

```python
def read_test(number):
    return pd.read_csv(f'./data/test/problem-A/A{number}.csv')

df_test_A1 = read_test(1)
print(df_test_A1.info())
df_test_A1.head()
```

```python
def find(CLUSTER, CPU_FREQ, DDR_FREQ):
    row = df_a_all.loc[
        (df_a_all['CLUSTER'] == CLUSTER) 
        & (df_a_all['CPU_FREQ'] == CPU_FREQ) 
        & (df_a_all['DDR_FREQ'] == DDR_FREQ)]
    sum = 0
    cnt = 0
    for i in row['IPS']:
        sum+=i
        cnt+=1
    return sum / cnt

print(find(0, 884000, 418000000))
out_A1 = df_test_A1
A1_ips = []
for index, row in df_test_A1.iterrows():
    CLUSTER = row['CLUSTER']
    CPU_FREQ = row['CPU_FREQ']
    DDR_FREQ = row['DDR_FREQ']
    A1_ips.append(find(CLUSTER, CPU_FREQ, DDR_FREQ))

out_A1['IPS'] = A1_ips

cols_send = [
    'CLUSTER',
    'CPU_FREQ',
    'DDR_FREQ',
    'IPS'
]

df_send_A1 = out_A1[cols_send]


print(df_send_A1.info())
df_send_A1.head()

df_send_A1.to_csv('./A1_simple1.csv',index=False)
```
